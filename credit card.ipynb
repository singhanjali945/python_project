```python
# import some important libraries
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset
df=pd.read_csv("creditcard.csv")
```

```python
# print information of dataset
df.info()
```

```python
# checking missing values
df.isnull().sum()
```

```python
# distribution of ligit transaction and fraudlant transaction
df["Class"].value_counts()
```

```python
# separation of data for the analysis
ligit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python

print(ligit.shape)
print(fraud.shape)
```

```python
#statistical measure of data
ligit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
df.groupby("Class").mean()
```

```python
legit_sample=ligit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model= LogisticRegression()
```

```python
# training the Logistic Regression Model with Trainig Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data: ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) 
```

```python
print('Accuracy on Test Data: ', test_data_accuracy)
```
